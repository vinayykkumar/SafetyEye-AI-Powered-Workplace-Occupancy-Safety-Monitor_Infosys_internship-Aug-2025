```python
from ultralytics import YOLO
import torch
from pathlib import Path

def main():
    """
    Trains the final, optimized YOLOv8 model using the best hyperparameters
    discovered during the tuning phase.
    """
    print("--- Training Final Model with Tuned Hyperparameters ---")
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Using device: {device}")

    # --- CONFIGURATION ---
    # 1. Specify the model size for your final model (e.g., yolov8m.pt)
    model_size = 'yolov8m.pt'

    # 2. Path to the YAML file with the best hyperparameters
    tuned_params_path = '../runs/detect/ppe_tuning2/best_hyperparameters.yaml'

    # 3. Path to the dataset's YAML file
    dataset_yaml_path = '../dataset/processed/data.yaml'
    project_name = '../runs/detect'
    run_name = 'ppe_final_tuned_model'

    # Check if the hyperparameters file exists
    if not Path(tuned_params_path).exists():
        print(f"Error: Hyperparameters file not found at '{tuned_params_path}'")
        print("Please run the tune.py script first to generate this file.")
        return

    model = YOLO(model_size)
    model.to(device)

    print(f"Starting final model training for 50 epochs...")
    print(f"Using tuned parameters from: {tuned_params_path}")
    
    # The 'cfg' argument overrides the default training
    # configurations with your optimized ones.
    results = model.train(
        # resume = True,
        cfg = tuned_params_path,
        data = dataset_yaml_path,
        epochs = 50,
        imgsz = 640,
        batch = -1,  # Auto-batch size for optimal VRAM usage
        name = 'ppe_final_tuned_model'
    )

    print("\nFinal tuned model training complete.")
    print(f"Best model is saved in '{project_name}/{run_name}/weights/best.pt'")

if __name__ == '__main__':
    main()
```
