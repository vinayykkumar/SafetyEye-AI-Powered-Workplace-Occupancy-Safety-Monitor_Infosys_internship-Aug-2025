```python
from ultralytics import YOLO
import torch
import pandas as pd

def main():
    print("Model Validation")
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Using device: {device}")

    model_path = '../runs/detect/ppe_final_tuned_model2/weights/best.pt'

    try:
        model = YOLO(model_path)
        model.to(device)
    except Exception as e:
        print(f"Error loading model: {e}")
        print(f"Please ensure the path '{model_path}' is correct.")
        return

    print("\nStarting model validation on the validation set...")
    # Validate the model on the dataset defined in the .yaml file
    # The 'split' argument specifies which dataset split to use (val, test, or train)
    metrics = model.val(split='val', plots=True)

    print("\n--- 📊 Overall Performance Metrics ---")
    # Overall box metrics
    map50_95 = metrics.box.map    # mAP50-95
    map50 = metrics.box.map50     # mAP50
    map75 = metrics.box.map75     # mAP75
    precision = metrics.box.p[0]  # Precision for the main class or overall
    recall = metrics.box.r[0]     # Recall for the main class or overall

    print(f"Mean Average Precision (mAP50-95): {map50_95:.4f}")
    print(f"Mean Average Precision (mAP50):    {map50:.4f}")
    print(f"Mean Average Precision (mAP75):    {map75:.4f}")
    print(f"Precision:                         {precision:.4f}")
    print(f"Recall:                            {recall:.4f}")

    print("\n\n--- 📋 Per-Class Accuracy Report ---")
    # Create a pandas DataFrame for a clean, tabular report
    class_names = model.names
    
    # Data for the report
    report_data = []
    for i, class_index in enumerate(metrics.box.ap_class_index):
        class_name = class_names[class_index]
        # Per-class metrics
        ap50 = metrics.box.ap50[i]
        ap = metrics.box.ap[i] # mAP50-95 for this class
        
        report_data.append([class_name, ap50, ap])

    # Create DataFrame
    df = pd.DataFrame(report_data, columns=['Class Name', 'mAP50', 'mAP50-95'])
    
    # Print the DataFrame without the index
    print(df.to_string(index=False))

    print("\n\nNote:")
    print(" - Precision: Of all the boxes the model drew, what percentage were correct?")
    print(" - Recall: Of all the actual objects present, what percentage did the model find?")
    print(" - mAP: The standard metric for object detection, balancing precision and recall.")
    print("\nValidation results and plots saved to:", metrics.save_dir)


if __name__ == '__main__':
    main()
```
