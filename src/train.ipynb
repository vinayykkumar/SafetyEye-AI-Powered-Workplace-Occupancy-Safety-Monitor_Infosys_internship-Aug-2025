```python
from ultralytics import YOLO
import torch

print("--- Milestone 2: Automated Hyperparameter Tuning ---")
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

model = YOLO('yolov8n.pt')
model.to(device)

print("Starting hyperparameter tuning...")

results = model.tune (
    # Core Training Parameters
    data = '../dataset/processed/data.yaml',
    epochs = 20,
    imgsz = 640,
    batch = -1, # Auto-batch size
    name = 'ppe_tuned_param',
    project = '../runs/detect',
    plots = False,        
    val = True,

    # Optimizer Parameters
    optimizer = 'AdamW',
    lr0 = 0.01055,          # Initial learning rate
    lrf = 0.01,             # Final learning rate factor (final_lr = lr0 * lrf)
    momentum = 0.937,       # AdamW beta1
    weight_decay = 0.0005,  # Weight decay regularization

    # Learning Rate Scheduler Parameters
    warmup_epochs = 3.0,
    warmup_momentum = 0.74549,

    # Loss Function Gains
    box = 7.5,              # Bounding box loss gain
    cls = 0.55967,          # Classification loss gain
    dfl = 1.58804,          # Distribution Focal Loss gain

    # Augmentation - Color Space
    hsv_h = 0.01473,        # Hue saturation value (Hue)
    hsv_s = 0.62684,        # Hue saturation value (Saturation)
    hsv_v = 0.33399,        # Hue saturation value (Value)
    bgr = 0.0,              # Blue-Green-Red channel shift

    # Augmentation - Geometric Transformations
    translate = 0.10187,    # Translation
    scale = 0.4235,         # Scaling (zoom)
    fliplr = 0.53616,       # Flip left-right

    # Augmentation - Advanced Methods
    mosaic = 0.85194,       # Mosaic augmentation probability
)
```
