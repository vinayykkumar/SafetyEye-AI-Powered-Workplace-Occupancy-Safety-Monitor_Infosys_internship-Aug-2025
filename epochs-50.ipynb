```python
# STEP 1: Install YOLOv8
!pip install -q ultralytics

# STEP 2: Import Library
from ultralytics import YOLO
import torch
torch.cuda.empty_cache()  # Clear GPU cache

# STEP 3: Load YOLOv8 Medium Model (Balanced: Speed + Accuracy)
model = YOLO("yolov8m.pt")  # You can also try 'yolov8s.pt' if memory issue continues

# STEP 4: Train the Model (Optimized for Kaggle GPU)
model.train(
    data="/kaggle/input/construction-site-safety-2/data.yaml",  # Path to dataset
    epochs=50,               # Reduced for faster training
    imgsz=640,               # Image size (balanced)
    batch=16,                # Reduce if still getting memory error
    optimizer="AdamW",       # Faster convergence
    lr0=0.003,               # Learning rate
    momentum=0.937,
    weight_decay=0.0005,

    # Augmentation settings (reduced for less memory usage)
    hsv_h=0.01487,
    hsv_s=0.62684,
    hsv_v=0.3339,
    translate=0.10187,
    scale=0.42305,
    mosaic=0.5,              # reduced
    mixup=0.0,               # off
    copy_paste=0.0,          # off

    patience=20              # early stopping
)

```
